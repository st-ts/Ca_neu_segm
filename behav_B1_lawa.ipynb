```python
# Create and display a FileChooser widget
fc = FileChooser('D:/behav/')
fc.filter_pattern = '*.csv'
fc.use_dir_icons = True
display(fc)
```

```python
file_path = fc.selected_path + '\\' + fc.selected_filename
data = pd.read_csv(file_path)
data_licks_L = pd.read_csv(file_path[0:-4] + '_t_licks_L.csv', header = None)
data_licks_R = pd.read_csv(file_path[0:-4] + '_t_licks_R.csv', header = None)
# Show the first few rows of the DataFrame
data = data[data['t_laser'] != 0]
# Step 1: Calculate the Differences
data['dur_trial'] = data['t_laser'].shift(-1) - data['t_laser']

# Step 2: Handle the Last Row
data['dur_trial'].fillna(20000, inplace=True)

data['end_trial'] = data['t_laser'].shift(-1) 
data['end_trial'].fillna( data['t_laser'][data.shape[0]-1] + 6000, inplace=True)
data
```

```python
def count_licks_in_trial(row):
    # Get licks within the start and end time of the trial
    licks_in_trial_R = len(data_licks_R[(data_licks_R[0] >= row['t_laser']) & 
                                 (data_licks_R[0] <= row['end_trial'])]) 
    licks_in_trial_L = len(data_licks_L[(data_licks_L[0] >= row['t_laser']) & 
                                 (data_licks_L[0] <= row['end_trial'])])
    return licks_in_trial_L + licks_in_trial_R


data['num_licks'] = data.apply(count_licks_in_trial, axis=1)
```

```python
data_las = data[data['laser']==1]
data_nolas = data[data['laser']==0]

results = pd.DataFrame()
results['port state'] = ['no water', 'water']
results['no_laser'] = [data_nolas[data_nolas['port']==1]['num_licks'].mean(), data_nolas[data_nolas['port']==2]['num_licks'].mean()]
results['laser'] = [data_las[data_las['port']==1]['num_licks'].mean(), data_las[data_las['port']==2]['num_licks'].mean()]
```
