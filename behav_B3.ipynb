```python
def count_licks_in_trial(row, data_licks):
    # Get licks within the start and end time of the trial
    licks_in_trial = len(data_licks[(data_licks[0] >= row['laser_ts']) & 
                                 (data_licks[0] <= row['laser_ts'] + 3000)]) 

    return licks_in_trial


data['num_licks_L'] = data.apply(lambda row: count_licks_in_trial(row, data_licks_L), axis=1)
data['num_licks_R'] = data.apply(lambda row: count_licks_in_trial(row, data_licks_R), axis=1)

data[data['laser_ts']!=0]
```

```python


# Assuming 'data', 'data_licks_L', and 'data_licks_R' are your DataFrames
# and 'laser' is the column with laser types in 'data'

# Combine left and right licks into one DataFrame
data_licks_L['lick_side'] = 'Left'
data_licks_R['lick_side'] = 'Right'
combined_licks = pd.concat([data_licks_L.rename(columns={0: 'lick_time'}), 
                            data_licks_R.rename(columns={0: 'lick_time'})])

# Merge lick data with laser data based on your time criteria
# Adjust the time criteria as needed
merged_data = pd.merge(data, combined_licks, how='left', 
                       left_on='laser_stim_ts', right_on='lick_time')

# Filter licks for each laser type and plot histograms
laser_types = data['laser'].unique()
for laser_type in laser_types:
    filtered_licks = merged_data[merged_data['laser'] == laser_type]['lick_time']
    plt.figure()
    plt.hist(filtered_licks, bins=20, alpha=0.7, label=f'Laser Type {laser_type}')
    plt.xlabel('Lick Time')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Lick Times for Laser Type {laser_type}')
    plt.legend()
    plt.show()
```
